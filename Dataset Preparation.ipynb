```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import time
import csv
import json
```

```python
df = pd.read_csv('majors-list.csv')
pd.set_option("display.max_rows", None)
df
```

```python
# Check Unique faculty in dataset
print(f"Unique Faculty:\n{df['Major_Category'].unique()}")
nan_values = df[df['Major_Category'].isna()]

# Locate the NaN values and drop it, Also drop Interdisciplinary to incerese consistency
print(f"\nLocated NaN Values\n{nan_values}\n")
df = df.drop(145)
df = df.drop(139)
df.reset_index(drop=True, inplace=True)

# Unique faculty after drop
unique_faculty = df['Major_Category'].unique()
print(f"\nUnique Faculty:\n{unique_faculty}")
```

```python
# Create Dictionary with keys name of faculty and values list of majors in that faculty
Faculty = {}
for x,y in zip(unique_faculty,unique_faculty):
    Faculty[f"{x}"] = df[df['Major_Category']==f"{y}"]['Major'].values
```

```python
# Manually Inputted Data
data = {"intents": [
        {"tag": "greeting",
         "patterns": ["Hi", "how are you", "Is anyone there?", "Hello", "Good day", "Whats up", "good morning", "good night", "good afternoon", "aloha"],
         "responses": ["Hello!", "Good to see you again!", "Hi there, how can I help?"],
        },
        {"tag": "goodbye",
         "patterns": ["cya", "see you later", "goodbye", "I am Leaving", "Have a Good day", "bye", "thankyou", "thanks"],
         "responses": ["Sad to see you go :(", "Talk to you later", "Goodbye!"],
        },
        {"tag": "name",
         "patterns": ["what is your name", "what should I call you", "whats your name?", "who are you?"],
         "responses": ["You can call me Aurora.", "I'm Aurora!", "Pleasure to meet you, I'm Aurora"],
        },
        {"tag": "pricing",
         "patterns": ["what is your rate", "How much for an hour", "Can you tell me about pricing?", "how much do you charge", "whats your price?"],
         "responses": ["Our Rate is Rp.10.000/hour", "You can talk to our mentor for Rp.10.000/hour", "Our mentor usually charge Rp.10.000/hour"],
        },
        {"tag": "time",
         "patterns": ["what time can i ask for a tutor", "When can i ask for a tutor", "Can i ask for a tutor in the afternoon"],
         "responses": ["You can call our tutor from 9am to 9pm, monday through saturday."],
        }
]}
```

```python
# Loop for each unique faculty
for item in unique_faculty:
    faculty = item
    majors = Faculty.get(f"{item}")
    lines1 = []
    # Loop for each majors in the faculty
    for majors_ in majors:
        question_format = [
            F"Can i know about {majors_.lower()}",
            F"Can you tell me about {majors_.lower()}",
            F"What about {majors_.lower()}",
            F"Do you mind telling me about {majors_.lower()}",
            F"Would you mind inform me about {majors_.lower()}"
        ]
        answer_format = [
            F"That majors is a part of Faculty of {faculty.lower()}",
            F"Alright, That majors is a part of {faculty.lower()}",
            F"That Majors is a part of {faculty.lower()}",
            F"Sure, That Majors is a part of {faculty.lower()}"
        ]
        # Loop through questin and answer format to create each sentences
        for question_format in (question_format):
            question = question_format
            lines1.append(question)
    # Create dictionary to convert into a json file
    tmp = {}
    tmp.update({
        "tag":f"{faculty.lower()}",
        "patterns": lines1,
        "responses": answer_format,
        })
    data["intents"].append(tmp)

with open('intent1.json', 'w') as fp:
    json.dump(data, fp)
```
